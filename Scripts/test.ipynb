```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
ratings = pd.read_csv('../data/ratings.csv')
books_data = pd.read_csv('../data/books.csv')
```

```python
ratings.head()
```

```python
books_data.head()
```

```python
#Look for NaN's
ratings.apply(lambda x: x.isnull().sum(),axis=0)
```

```python
books = pd.DataFrame(books_data, columns=['book_id', 'authors', 'title', 'average_rating','ratings_count', 'image_url'])
books.head()
```

```python
books_info = pd.merge(books, ratings, on='book_id')
books_info.head()
```

```python
book_rating_matrix = pd.pivot_table(books_info, index='user_id', values='rating', columns='book_id', fill_value=0)
book_rating_matrix
```

```python
input_book = 1
book_entered = book_rating_matrix[input_book]
similar_to_book_entered = book_rating_matrix.corrwith(book_entered)
similar_to_book_entered = similar_to_book_entered.sort_values(ascending=False)
similar_to_book_entered = list(similar_to_book_entered[1:6])
similar_to_book_entered
```

```python
book_entered = books_data.loc[books_data['id'] == input_book]
print("Book entered is", book_entered['original_title'])
```
